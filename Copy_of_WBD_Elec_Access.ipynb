# **Task 1. Ingest and manipulate the data using pandas dataframes.**
Your program should include a function which takes a filename as argument, reads a dataframe in Worldbank format and returns two dataframes: one with years as columns and one with
countries as columns. Do not forget to clean the transposed dataframe.

```python
#Creating python Function 
import pandas as pd

def read_world_bank_data(filename):
    # read in the CSV file
    df = pd.read_csv(filename)

    # drop irrelevant columns
    df = df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'])

    # set the index to be the country name
    df = df.set_index('Country Name')

    # pivot the dataframe to get years as columns
    df_years = df.transpose()

    # clean the dataframes
    df_years.index.name = 'Year'
    df_years.columns.name = None

    # remove any rows or columns with all NaN values
    df_years = df_years.dropna(axis=1, how='all')

    # pivot the dataframe again to get countries as columns
    df_countries = df_years.transpose()

    return df_years, df_countries
```

# **Task 2.  Explore the statistical properties**

```python
import pandas as pd

# Read in the data
df = pd.read_csv('/content/API_EG.ELC.ACCS.ZS_DS2_en_csv_v2_4902219.csv')

# Select the relevant columns
cols = ['Country Name', 'Indicator Name', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019','2020']
df = df[cols]
df
```

Next, let's calculate some summary statistics for a few countries:

```python
# Calculate summary statistics for a few countries
countries = ['United States', 'China', 'India', 'Germany']
for country in countries:
    subset = df[df['Country Name'] == country].iloc[:, 2:]
    print(f'Summary statistics for {country}:')
    print(subset.describe())
    print('\n')
```

This will print out summary statistics (mean, standard deviation, min, max, and quartiles) for each of the specified countries.

You can also calculate summary statistics for the entire world by selecting all rows and calculating the mean, standard deviation, etc.:

```python
# Calculate summary statistics for the entire world
world = df.iloc[:, 2:]
print('Summary statistics for the world:')
print(world.describe())
```

In addition to using the describe() method, you can also calculate other statistics using pandas functions, such as the mean, median, and standard deviation:

```python
# Calculate the mean, median, and standard deviation for a few countries
for country in countries:
    subset = df[df['Country Name'] == country].iloc[:, 2:]
    print(f'Statistics for {country}:')
    print(f'Mean: {subset.mean()}')
    print(f'Median: {subset.median()}')
    print(f'Standard deviation: {subset.std()}')
    print('\n')
```
